```python
p_price_df[0].head()
```

```python

i=0

# dict to store adjclose of each ticker temporary
temp_price_df_adj={}

# dict to store log return of each ticker temporary
temp_log_ret={}

# list storing log return of all tickers
p_log_ret=[]


while i < len(p_tickers):

    temp_price_df_adj = p_price_df[i]['adjclose']
    
    
    # This will calculate log returns for the series
    temp_log_ret = np.log(temp_price_df_adj) - np.log(temp_price_df_adj.shift(1))
    
    p_log_ret.append(temp_log_ret)
    i+=1
```

```python
#convert to a single pd.dataframe
p_pd_log_ret = pd.DataFrame()

for i,ticker in enumerate(p_tickers):
    p_pd_log_ret[ticker] = p_log_ret[i]
```

```python
p_pd_log_ret
```

```python
#Initialize a list for weighting
p_weighting = [0.25,0.25,0.25,0.25]
```

```python
#convert the weight from list to np array 
p_portfolio_weights = np.array(p_weighting)
```

```python
p_portfolio_weights
```

```python
p_pd_log_ret.mean()
```

```python
expected_return = np.sum((p_pd_log_ret.mean() * p_portfolio_weights))
```

```python
# Annualized return
p_ann_return = expected_return*252
```

```python
p_ann_return 
```

```python
p_pd_log_ret.cov()
```

```python
#calculate variance part 1
#(weight_1)^2*var_1 + (weight_2)^2*var_2 + ... +  (weight_4)^2*var_4

p_port_var_A = 0

for i,ticker in enumerate(p_tickers):
    p_port_var_A += p_portfolio_weights[i]**2 * p_pd_log_ret[ticker].var() 
```

```python
p_port_var_A
```

```python
#calculate variance part 2
#2*weight_1*weight_2*cov_1&2 +  2*weight_1*weight_3*cov_1&3 + ... +  2*weight_3*weight_4*cov_3&4

p_port_var_B = 0
combination_list = []
combination = ''
combination_reverse = ''

for x,ticker in enumerate(p_tickers):
    for y,ticker_2 in enumerate(p_tickers):
        if ticker == ticker_2:
            pass
        else:
            
            combination = ticker + "*" + ticker_2
            combination_reverse = ticker_2 + "*" + ticker
            
            if (combination not in combination_list) and (combination_reverse not in combination_list):
                p_port_var_B += 2*p_portfolio_weights[x]*p_portfolio_weights[y]*p_pd_log_ret.cov().loc[ticker,ticker_2]
                combination_list.append(combination)
            else:
                pass
```

```python
p_port_var_B
```

```python
# variance
p_port_var = p_port_var_A + p_port_var_B
```

```python
p_port_var
```

```python
# Annual variance
print (p_port_var*252)
```

```python
# Standard deviation
p_port_std = np.sqrt(p_port_var*252)
print(p_port_std)
```

```python
#sharpe ratio

p_sharpe = (p_ann_return - 0)/p_port_std
print(p_sharpe)
```

```python
# add sum of portfolio into the dataframe
p_pd_log_ret.loc[:,'Portfolio']= p_pd_log_ret.sum(axis=1)
```

```python
# DIA = Dow Jones index. This DIA is the portfolio.
yahoo_financials = YahooFinancials('DIA')
    
# Get historical market_return
market_return = yahoo_financials.get_historical_price_data('2022-01-01', '2022-10-31', 'daily')


market_return_dict2 = pd.DataFrame(market_return['DIA']['prices'])
market_return_dict2.index = pd.to_datetime(market_return_dict2['formatted_date'])

market_return_dict2 = market_return_dict2['close']
market_return_dict2 = np.log(market_return_dict2/market_return_dict2.shift(1))
market_return_dict2 = market_return_dict2.drop(market_return_dict2.index[0])

```

```python
market_return_dict2
```

```python
# Add the market return into the dataframe
p_pd_log_ret['DIA'] = market_return_dict2
```

```python
p_pd_log_ret = p_pd_log_ret.drop(p_pd_log_ret.index[0])
```

```python
p_pd_log_ret
```

```python
# import statsmodels API
import statsmodels.api as sm

# X is the market index return
X = p_pd_log_ret['DIA']

# y is the portfolio return
y = p_pd_log_ret['Portfolio']

# Add a constant (alpha) in the regression model
# An intercept is not included by default and should be added by the user
X1 = sm.add_constant(X)

# Regression model
# OLS: Ordinary Least Squares
model = sm.OLS(y,X1)

# Fit the model and print results
p_results = model.fit()
print(p_results.summary())
```

```python
print(p_results.params)
```

```python
# beta
p_beta = p_results.params[1]
print('Beta is ' + str(p_beta))
```

```python
print('R2: ', p_results.rsquared)
```

```python
#treynor ratio
p_treynor = (p_ann_return - 0)/p_beta
print(p_treynor)
```

```python
# annual market return
p_ann_market_return =  market_return_dict2.mean() *252

# alpha = portfolio return - risk free rate - beta(market return - risk-free rate)
p_alpha = p_ann_return - 0 - p_beta*(p_ann_market_return - 0)
print(p_alpha)
```
